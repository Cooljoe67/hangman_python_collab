```python

# hangman, when dead step 12
h_man = [[
    " /MMM\\   "
   ," |x x|   "
   ," \\ o /   "
   ,"   T     "
   ," /WWW\\   "
   ,"/ WWW \\  "
   ,"  I I    "
   ,"  L L    "
   ,"         "
   ]]

# hangman, lower left leg missing
h_man.insert(1,h_man[0].copy())
h_man[0][1] = " |o o|   "
h_man[0][2] = " \\ - /   "
h_man[0][7] = "    L    "

# hangman, comlete left leg missing
h_man.insert(1,h_man[0].copy())
h_man[0][6] = "    I    "

# hangman, lower right leg missing
h_man.insert(1,h_man[0].copy())
h_man[0][7] = "         "

# hangman, comlete right leg missing
h_man.insert(1,h_man[0].copy())
h_man[0][6] = "         "

# hangman, lower left arm missing
h_man.insert(1,h_man[0].copy())
h_man[0][5] = "  WWW \\  "

# hangman, complete left arm missing
h_man.insert(1,h_man[0].copy())
h_man[0][4] = "  WWW\\   "

# hangman, lower right arm missing
h_man.insert(1,h_man[0].copy())
h_man[0][5] = "  WWW    "

# hangman, complete left arm missing
h_man.insert(1,h_man[0].copy())
h_man[0][4] = "  WWW    "

# hangman, lower body missing, still smiling
h_man.insert(1,h_man[0].copy())
h_man[0][5] = "         "
h_man[0][2] = " \\ u /   "

# hangman, complete body missing
h_man.insert(1,h_man[0].copy())
h_man[0][4] = "         "

# hangman, neck missing
h_man.insert(1,h_man[0].copy())
h_man[0][3] = "         "

# hangman, empty
h_man.insert(1,h_man[0].copy())
h_man[0][0] = "         "
h_man[0][1] = "         "
h_man[0][2] = "         "

# hangman, happy
h_man.append([

    "         "
   ," /MMM\\   "
   ," |o o|   "
   ," \\ u /   "
   ,"\\  T  /  "
   ," -WWW-   "
   ,"  WWW    "
   ,"  I I    "
   ,"  L L    "
   ])

#for mn in h_man:
#  for ln in mn:
#    print(ln)
#  print()
#  print("------------------")
#  print()

letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
used    = "__________________________"

# words to be used as a follback if api call fails
random_words = ["velocity", "labyrinth", "ethereal", "cascade", "nebula", "quixotic", "serendipity", "halcyon", "ephemeral", "resilience"]

import requests

api_url = 'https://api.api-ninjas.com/v2/randomword'
response = requests.get(api_url, headers={'X-Api-Key': 'DFQvmocca69fudiiizHfrA==UKBogXPXiYjbUywt'})
if response.status_code == requests.codes.ok:
    secret=response.text[2:-2].upper()
else:
    import random
    secret=random_words[random.randint(0, 9)].upper()


# ---- this is basically the game. it's called recursivley
def game_round(stage,level):
  global used
  # print(secret)

  print()
  print("---------------------------------------------------")
  print("XXXXXXXXXX")
  print("X        |")
  print(f"X     {h_man[stage][0]}\t{used[0:5]}")
  print(f"X     {h_man[stage][1]}\t{used[5:10]}")
  print(f"X     {h_man[stage][2]}\t{used[10:15]}")
  print(f"X     {h_man[stage][3]}\t{used[15:20]}")
  print(f"X     {h_man[stage][4]}\t{used[20:26]}")
  print(f"X     {h_man[stage][5]}")
  print(f"X     {h_man[stage][6]}")
  print(f"X     {h_man[stage][7]}")
  print(f"X     {h_man[stage][8]}")
  print()
  print()

  # print the secret word with only the found chars showing
  # (replace not used chars with '_')
  mymapping = str.maketrans(letters, used)
  for j in secret.translate(mymapping):
    print(j+" ",end=" ")

  #stages 13 (lost) or 14 (won) dont need any of this, it's already over
  if stage < 12:
    guess=input("\nLetter or solution ('bye' to give up) : ").upper()
    #correct guess
    if guess==secret:
      print("GREAT! You got the word!!!!!")
      game_round(13, level)
    elif guess=="BYE":
      game_round(12, level)
    #word was entered, but the worn one
    elif len(guess)>1 and secret != guess:
      print("You guessed wrong!")
      game_round(stage + level,level)
    #letter was entered, but one already in the used letter list
    elif len(guess)==1 and used.find(guess) != -1:
      print("You alredy used that letter, try again!")
      game_round(stage, level)
    #letter was entered, but an illegal one
    elif len(guess)==1 and (guess < "A" or guess > "Z"):
      print("Only letters fom A to Z please, try again!")
      game_round(stage,level)
    #letter was entered, but one not one that's in the secret word
    elif len(guess)==1 and secret.find(guess) == -1:
      print("That letter is NOT in the secret word!")
      if stage+level == 12: print("YOU LOST!!!!")
      used = used[:ord(guess)-65] + guess + used[ord(guess)-64:]
      game_round(stage + level,level)
    elif len(guess)==1:
      used = used[:ord(guess)-65] + guess + used[ord(guess)-64:]
      mymapping = str.maketrans(letters, used)
      if secret.translate(mymapping) == secret:
        print("GREAT! You got the word!!!!!")
        game_round(13,level )
      else:
        print("Super! You got one!")
        game_round(stage, level)


  if stage==12:
    print("\nToo bad, the word was:" , secret)
  return


# ------ Everything set, now do it.... ------
lvl=0
while lvl != 1 and lvl !=2:
  lvl=int(input("Dificulty (1/2)? "))


game_round(0,lvl)

```
